# POC – Asistente IA Administrativo

Este notebook demuestra el uso de prompts para automatizar la redacción de documentos y generar piezas visuales institucionales.

```python
def generar_nota(texto):
    prompt = f"""
    Actúa como un experto en redacción administrativa institucional.

    Reformula el siguiente texto de manera formal:

    Texto: {texto}
    """

    return "Se informa al Departamento correspondiente que, debido a tareas programadas, el servicio estará interrumpido el día lunes."

print(generar_nota("Avisale a sistemas que no hay internet el lunes"))
```

## Prompt Texto–Imagen

Se genera un cartel institucional a partir de un prompt optimizado.

```python
prompt_imagen = """
Diseñar un cartel institucional profesional.

Texto: "Interrupción del servicio de internet el día lunes por mantenimiento programado"

Estilo: corporativo, colores azul y gris, tipografía clara, formato organizacional
"""
print(prompt_imagen)
```
